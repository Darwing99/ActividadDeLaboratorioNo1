## Parte II: Estadísticas Descriptivas

```python
salario_max   = df["Salario"].max()
salario_min   = df["Salario"].min()
salario_prom  = df["Salario"].mean()
mediana       = df["Salario"].median()
desv_std      = df["Salario"].std()

print("=" * 40)
print("  ESTADÍSTICAS DESCRIPTIVAS DE SALARIO")
print("=" * 40)
print(f"  Salario Máximo    : ${salario_max:>12,.2f}")
print(f"  Salario Mínimo    : ${salario_min:>12,.2f}")
print(f"  Salario Promedio  : ${salario_prom:>12,.2f}")
print(f"  Mediana Salarial  : ${mediana:>12,.2f}")
print(f"  Desviación Estánd.: ${desv_std:>12,.2f}")
print("=" * 40)
```

---
## Parte III: Análisis Organizacional

```python
# Cantidad de cargos y países
num_cargos  = df["Cargo"].nunique()
num_paises  = df["País"].nunique()

print(f"Cantidad de cargos existentes : {num_cargos}")
print(f"Cantidad de países representados: {num_paises}")
print()
print("Listado de cargos:")
for cargo in sorted(df["Cargo"].unique()):
    print(f"  - {cargo}")
print()
print("Listado de países:")
for pais in sorted(df["País"].unique()):
    print(f"  - {pais}")
```

```python
# Cargo con salario promedio más alto
prom_por_cargo = df.groupby("Cargo")["Salario"].mean().sort_values(ascending=False)

cargo_max_nombre = prom_por_cargo.idxmax()
cargo_max_valor  = prom_por_cargo.max()

print(f"Cargo con salario promedio más ALTO:")
print(f"  {cargo_max_nombre}  →  ${cargo_max_valor:,.2f}")
print()
print("Promedio salarial por cargo (mayor a menor):")
for cargo, val in prom_por_cargo.items():
    print(f"  {cargo:<50} ${val:>10,.2f}")
```

```python
# Cargo con salario promedio más bajo
cargo_min_nombre = prom_por_cargo.idxmin()
cargo_min_valor  = prom_por_cargo.min()

print(f"Cargo con salario promedio más BAJO:")
print(f"  {cargo_min_nombre}  →  ${cargo_min_valor:,.2f}")
```

```python
# País con el salario promedio más alto
prom_por_pais    = df.groupby("País")["Salario"].mean().sort_values(ascending=False)
pais_max_nombre  = prom_por_pais.idxmax()
pais_max_valor   = prom_por_pais.max()

print(f"País con salario promedio más ALTO:")
print(f"  {pais_max_nombre}  →  ${pais_max_valor:,.2f}")
print()
print("Promedio salarial por país (mayor a menor):")
for pais, val in prom_por_pais.items():
    print(f"  {pais:<20} ${val:>10,.2f}")
```

## Parte IV: Visualización

```python
# Gráfico de barras — Salario promedio por cargo
prom_cargo_sorted = prom_por_cargo.sort_values()
colores_cargo = ["#ff6b6b" if v == prom_cargo_sorted.max() else "#4fc3f7"
                 for v in prom_cargo_sorted]

fig, ax = plt.subplots(figsize=(13, 6))
bars = ax.barh(prom_cargo_sorted.index, prom_cargo_sorted.values,
               color=colores_cargo, edgecolor="#0f0f0f")

for bar, val in zip(bars, prom_cargo_sorted.values):
    ax.text(val + 800, bar.get_y() + bar.get_height() / 2,
            f"${val:,.0f}", va="center", fontsize=9, color="white")

ax.set_title("Gráfico de Barras — Salario Promedio por Cargo")
ax.set_xlabel("Salario Promedio (USD)")
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"${x:,.0f}"))
ax.grid(True, axis="x")

plt.tight_layout()
plt.savefig(f"{FIGURES_DIR}/01_barras_salario_por_cargo.png", dpi=150, bbox_inches="tight")
plt.show()
print("Gráfico guardado.")
```

```python
# Histograma — Distribución salarial
fig, ax = plt.subplots(figsize=(12, 5))

ax.hist(df["Salario"], bins=40, color="#4fc3f7", edgecolor="#0f0f0f", alpha=0.85)
ax.axvline(salario_prom,  color="#ff6b6b", linewidth=2, linestyle="--",
           label=f"Promedio: ${salario_prom:,.0f}")
ax.axvline(mediana, color="#69ff47", linewidth=2, linestyle="--",
           label=f"Mediana:  ${mediana:,.0f}")

ax.set_title("Histograma — Distribución de Salarios Tecnológicos")
ax.set_xlabel("Salario (USD)")
ax.set_ylabel("Cantidad de Empleados")
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"${x:,.0f}"))
ax.legend()
ax.grid(True)

plt.tight_layout()
plt.savefig(f"{FIGURES_DIR}/02_histograma_salarios.png", dpi=150, bbox_inches="tight")
plt.show()
print("Gráfico guardado.")
```

```python
# Gráfico circular — Empleados por país
empleados_pais = df["País"].value_counts()

# Agrupa países con menos del 3 % en "Otros" para mejor legibilidad
umbral = 0.03 * empleados_pais.sum()
principales = empleados_pais[empleados_pais >= umbral]
otros_total  = empleados_pais[empleados_pais < umbral].sum()
if otros_total > 0:
    principales["Otros"] = otros_total

colores_pie = plt.cm.tab20.colors[:len(principales)]
explode     = [0.04] * len(principales)

fig, ax = plt.subplots(figsize=(10, 7))
fig.patch.set_facecolor("#0f0f0f")
ax.set_facecolor("#0f0f0f")

wedges, texts, autotexts = ax.pie(
    principales.values,
    labels=principales.index,
    autopct="%1.1f%%",
    colors=colores_pie,
    explode=explode,
    startangle=140,
    pctdistance=0.82,
    textprops={"color": "white", "fontsize": 9},
)
for at in autotexts:
    at.set_fontsize(8)
    at.set_color("white")

ax.set_title("Gráfico Circular — Distribución de Empleados por País",
             fontsize=13, color="white", pad=15)

plt.tight_layout()
plt.savefig(f"{FIGURES_DIR}/03_circular_empleados_por_pais.png", dpi=150, bbox_inches="tight")
plt.show()
print("Gráfico guardado.")
```

## Parte V: Interpretación

### Pregunta 1: ¿Qué hallazgo considera más importante?

El hallazgo más importante es la **marcada concentración del talento y las mejores remuneraciones en dos países**: Estados Unidos y Canadá, cuyos salarios promedio ($113,484 y $100,228 respectivamente) casi duplican el promedio global del dataset ($63,541). Esto evidencia una brecha geográfica severa que impacta directamente la equidad salarial dentro de la organización y puede dificultar la retención del personal ubicado en Centroamérica y Latinoamérica, donde los salarios promedios oscilan entre $47,850 (Nicaragua) y $77,164 (España).

Un hallazgo complementario relevante es la asimetría entre el rol con mayor y menor pago: el **Arquitecto de Datos** gana en promedio $117,761, mientras que el **Analista de Datos Junior** recibe apenas $23,050 — una diferencia de más de cinco veces, lo que señala una pirámide salarial muy pronunciada.

### Pregunta 2: ¿Existe una diferencia significativa entre los salarios máximos y mínimos?

Sí, la diferencia es extremadamente significativa. El salario máximo registrado es de **$268,000** (Arquitecto de Datos en Estados Unidos), mientras que el mínimo es de **$13,000** (Analista de Datos Junior en Nicaragua/Honduras), lo que representa una brecha de **$255,000** — casi **21 veces** el salario más bajo.

Además, la desviación estándar de $31,586 sobre un promedio de $63,541 equivale a un **coeficiente de variación del ~49.7 %**, lo que confirma que la dispersión salarial es muy alta. El histograma de distribución muestra una curva claramente sesgada hacia la derecha (*right-skewed*), con la mayoría de empleados concentrados en rangos bajos ($13,000–$60,000) y una cola larga de salarios altos, lo que distorsiona el promedio hacia arriba respecto a la mediana ($59,000).

### Pregunta 3: ¿Qué cargo parece ser el mejor remunerado?

El cargo mejor remunerado es **Arquitecto de Datos**, con un salario promedio de **$117,761** y un salario máximo individual de $268,000. Le siguen el **Ingeniero Machine Learning** ($106,266) y el **Científico de Datos** ($92,245).

Estos tres cargos comparten características comunes: alta especialización técnica, demanda en el mercado global y dominancia en países con economías fuertes como Estados Unidos y Canadá. La naturaleza estratégica de estos roles — diseño de infraestructura de datos, modelos predictivos y análisis avanzado — justifica la prima salarial significativa que reciben frente a los roles más operativos.

### Pregunta 4: ¿Qué país presenta mejores condiciones salariales?

**Estados Unidos** presenta las mejores condiciones salariales con un promedio de **$113,484**, seguido de **Canadá** con $100,228. Ambos países superan considerablemente al resto de la muestra.

Sin embargo, es importante contextualizar este resultado: el costo de vida en Estados Unidos y Canadá también es significativamente mayor. Para una evaluación de condiciones salariales reales (poder adquisitivo), sería necesario incorporar índices de paridad de poder adquisitivo (PPP). En términos nominales, no obstante, **Estados Unidos es el líder indiscutido** del dataset.

Dentro de la región latinoamericana, **España** lidera con $77,164, seguida por **Panamá** ($70,421) y **Chile** ($66,411), que ofrecen condiciones más competitivas que el promedio regional.

### Pregunta 5: ¿Qué recomendaciones realizaría a la Gerencia de Recursos Humanos?

Con base en el análisis realizado, se presentan las siguientes recomendaciones:

1. **Revisar la equidad salarial geográfica**: La brecha entre países como Estados Unidos ($113,484) y Nicaragua ($47,850) es de más de 2.3 veces. Se recomienda establecer bandas salariales regionales que contemplen el costo de vida local pero que también garanticen condiciones competitivas para retener talento en mercados emergentes.

2. **Invertir en planes de carrera para roles junior**: El salario del Analista de Datos Junior ($23,050 promedio) es significativamente bajo comparado con el resto. Crear rutas de ascenso claras y capacitaciones técnicas aumentaría la retención y motivación en los niveles de entrada.

3. **Fortalecer la captación de Arquitectos de Datos e Ingenieros ML**: Dado que son los perfiles mejor remunerados y más escasos, la organización debe desarrollar estrategias de atracción competitiva (salario, beneficios, trabajo remoto, formación) para evitar la fuga de talento hacia competidores o mercados internacionales.

4. **Analizar la distribución por género y antigüedad**: El dataset actual no incluye variables de género ni años de experiencia. Incorporar estas dimensiones en futuras recolecciones de datos permitirá detectar posibles brechas salariales injustificadas y tomar decisiones más informadas.

5. **Considerar incentivos no salariales en países de menor remuneración**: Para los países centroamericanos (Honduras, Nicaragua, Guatemala), donde los salarios promedio son los más bajos del dataset, complementar la oferta salarial con beneficios adicionales (seguro médico, trabajo remoto, bonificaciones por desempeño) puede mejorar la competitividad sin impactar significativamente la estructura de costos.
