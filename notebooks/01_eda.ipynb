```python
# ── Paso 19: Configuración global de gráficas ─────────────────────────────────
import os

plt.rcParams.update({
    "figure.facecolor": "#0f0f0f",
    "axes.facecolor":   "#1a1a2e",
    "axes.edgecolor":   "#444",
    "axes.labelcolor":  "white",
    "axes.titlesize":   13,
    "axes.titlecolor":  "white",
    "xtick.color":      "white",
    "ytick.color":      "white",
    "text.color":       "white",
    "grid.color":       "#333",
    "grid.linestyle":   "--",
    "grid.alpha":       0.5,
})

FIGURES_DIR = "reports/figures"
os.makedirs(FIGURES_DIR, exist_ok=True)
print("Configuración lista. Figuras se guardarán en:", FIGURES_DIR)
```

```python
# ── Paso 20: Distribución de Salarios (Histograma) ────────────────────────────
fig, ax = plt.subplots(figsize=(12, 5))

ax.hist(df["Salario"], bins=40, color="#4fc3f7", edgecolor="#0f0f0f", alpha=0.85)
ax.axvline(df["Salario"].mean(),   color="#ff6b6b", linewidth=2, linestyle="--", label=f'Promedio: ${df["Salario"].mean():,.0f}')
ax.axvline(df["Salario"].median(), color="#69ff47", linewidth=2, linestyle="--", label=f'Mediana:  ${df["Salario"].median():,.0f}')

ax.set_title("Distribución de Salarios Tecnológicos")
ax.set_xlabel("Salario (USD)")
ax.set_ylabel("Cantidad de Empleados")
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"${x:,.0f}"))
ax.legend()
ax.grid(True)

plt.tight_layout()
plt.savefig(f"{FIGURES_DIR}/01_distribucion_salarios.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
# ── Paso 21: Salario Promedio por Cargo (Barras Horizontales) ─────────────────
salario_por_cargo = df.groupby("Cargo")["Salario"].mean().sort_values()

colores = ["#ff6b6b" if v == salario_por_cargo.max() else "#4fc3f7" for v in salario_por_cargo]

fig, ax = plt.subplots(figsize=(12, 6))
bars = ax.barh(salario_por_cargo.index, salario_por_cargo.values, color=colores, edgecolor="#0f0f0f")

for bar, val in zip(bars, salario_por_cargo.values):
    ax.text(val + 800, bar.get_y() + bar.get_height() / 2,
            f"${val:,.0f}", va="center", fontsize=9, color="white")

ax.set_title("Salario Promedio por Cargo")
ax.set_xlabel("Salario Promedio (USD)")
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"${x:,.0f}"))
ax.grid(True, axis="x")

plt.tight_layout()
plt.savefig(f"{FIGURES_DIR}/02_salario_por_cargo.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
# ── Paso 22: Salario Promedio por País (Barras Horizontales) ──────────────────
salario_por_pais = df.groupby("País")["Salario"].mean().sort_values()

colores_pais = ["#ff6b6b" if v == salario_por_pais.max() else "#a29bfe" for v in salario_por_pais]

fig, ax = plt.subplots(figsize=(12, 6))
bars = ax.barh(salario_por_pais.index, salario_por_pais.values, color=colores_pais, edgecolor="#0f0f0f")

for bar, val in zip(bars, salario_por_pais.values):
    ax.text(val + 500, bar.get_y() + bar.get_height() / 2,
            f"${val:,.0f}", va="center", fontsize=9, color="white")

ax.set_title("Salario Promedio por País")
ax.set_xlabel("Salario Promedio (USD)")
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"${x:,.0f}"))
ax.grid(True, axis="x")

plt.tight_layout()
plt.savefig(f"{FIGURES_DIR}/03_salario_por_pais.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
# ── Paso 23: Empleados por País (Barras Verticales) ───────────────────────────
empleados_pais = df["País"].value_counts().sort_values(ascending=False)

fig, ax = plt.subplots(figsize=(13, 5))
bars = ax.bar(empleados_pais.index, empleados_pais.values, color="#55efc4", edgecolor="#0f0f0f")

for bar, val in zip(bars, empleados_pais.values):
    ax.text(bar.get_x() + bar.get_width() / 2, val + 5,
            str(val), ha="center", va="bottom", fontsize=9, color="white")

ax.set_title("Cantidad de Empleados por País")
ax.set_xlabel("País")
ax.set_ylabel("Cantidad de Empleados")
ax.tick_params(axis="x", rotation=35)
ax.grid(True, axis="y")

plt.tight_layout()
plt.savefig(f"{FIGURES_DIR}/04_empleados_por_pais.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
# ── Paso 24: Boxplot de Salarios por Cargo ────────────────────────────────────
orden_cargo = df.groupby("Cargo")["Salario"].median().sort_values().index.tolist()
grupos = [df[df["Cargo"] == c]["Salario"].values for c in orden_cargo]

fig, ax = plt.subplots(figsize=(13, 6))
bp = ax.boxplot(grupos, vert=False, patch_artist=True, notch=False,
                medianprops=dict(color="#ff6b6b", linewidth=2),
                whiskerprops=dict(color="#aaa"),
                capprops=dict(color="#aaa"),
                flierprops=dict(marker="o", color="#fdcb6e", alpha=0.4, markersize=4))

colors = plt.cm.cool([i / len(grupos) for i in range(len(grupos))])
for patch, color in zip(bp["boxes"], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

ax.set_yticks(range(1, len(orden_cargo) + 1))
ax.set_yticklabels(orden_cargo, fontsize=9)
ax.set_title("Distribución de Salarios por Cargo (Boxplot)")
ax.set_xlabel("Salario (USD)")
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"${x:,.0f}"))
ax.grid(True, axis="x")

plt.tight_layout()
plt.savefig(f"{FIGURES_DIR}/05_boxplot_cargo.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
# ── Paso 25: Dashboard Resumen (4 gráficas en una figura) ─────────────────────
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
fig.suptitle("Dashboard — Análisis de Salarios Tecnológicos", fontsize=15, fontweight="bold")

# 1) Histograma
ax = axes[0, 0]
ax.hist(df["Salario"], bins=35, color="#4fc3f7", edgecolor="#0f0f0f", alpha=0.85)
ax.axvline(df["Salario"].mean(),   color="#ff6b6b", linewidth=1.5, linestyle="--", label="Promedio")
ax.axvline(df["Salario"].median(), color="#69ff47", linewidth=1.5, linestyle="--", label="Mediana")
ax.set_title("Distribución de Salarios")
ax.set_xlabel("Salario (USD)")
ax.set_ylabel("Frecuencia")
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"${x/1000:.0f}k"))
ax.legend(fontsize=8)
ax.grid(True)

# 2) Salario promedio por cargo
ax = axes[0, 1]
sc = df.groupby("Cargo")["Salario"].mean().sort_values()
ax.barh(sc.index, sc.values, color="#a29bfe", edgecolor="#0f0f0f")
ax.set_title("Salario Promedio por Cargo")
ax.set_xlabel("USD")
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"${x/1000:.0f}k"))
ax.grid(True, axis="x")

# 3) Empleados por país
ax = axes[1, 0]
ep = df["País"].value_counts()
ax.bar(ep.index, ep.values, color="#55efc4", edgecolor="#0f0f0f")
ax.set_title("Empleados por País")
ax.set_ylabel("Cantidad")
ax.tick_params(axis="x", rotation=45)
ax.grid(True, axis="y")

# 4) Salario promedio por país
ax = axes[1, 1]
sp = df.groupby("País")["Salario"].mean().sort_values()
ax.barh(sp.index, sp.values, color="#fd79a8", edgecolor="#0f0f0f")
ax.set_title("Salario Promedio por País")
ax.set_xlabel("USD")
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"${x/1000:.0f}k"))
ax.grid(True, axis="x")

plt.tight_layout()
plt.savefig(f"{FIGURES_DIR}/06_dashboard_resumen.png", dpi=150, bbox_inches="tight")
plt.show()
print("Todas las gráficas guardadas en:", FIGURES_DIR)
```
